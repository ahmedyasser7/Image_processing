```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from scipy.ndimage import generic_filter
```

```python
image_path = 'ahmedyassertaha.jpg' 
image = cv2.imread(image_path)
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
gray = gray.astype(np.uint8)
```

```python
plt.figure(figsize=(15, 15))

plt.subplot(3, 3, 2)
plt.imshow(gray, cmap='gray')
plt.title('Original Grayscale Image')
plt.axis('off')


kernel = np.ones((3,3), np.float32)/9


avg_zero = cv2.filter2D(gray, -1, kernel, borderType=cv2.BORDER_CONSTANT)
plt.subplot(3, 3, 4)
plt.imshow(avg_zero, cmap='gray')
plt.title('Average Filter (Zero Padding)')
plt.axis('off')


avg_ignore = cv2.filter2D(gray, -1, kernel, borderType=cv2.BORDER_ISOLATED)
plt.subplot(3, 3, 5)
plt.imshow(avg_ignore, cmap='gray')
plt.title('Average Filter (Ignore Edges)')
plt.axis('off')

def min_filter(region):
    return np.min(region)

def max_filter(region):
    return np.max(region)

def midpoint_filter(region):
    return (np.max(region) + np.min(region)) / 2

min_filtered = generic_filter(gray, min_filter, size=3)
plt.subplot(3, 3, 6)
plt.imshow(min_filtered, cmap='gray', vmin=0, vmax=255) 
plt.title('Min Filter (3x3)')
plt.axis('off')

max_filtered = generic_filter(gray, max_filter, size=3)
plt.subplot(3, 3, 7)
plt.imshow(max_filtered, cmap='gray', vmin=0, vmax=255)
plt.title('Max Filter (3x3)')
plt.axis('off')

median_filtered = cv2.medianBlur(gray, 3)
plt.subplot(3, 3, 8)
plt.imshow(median_filtered, cmap='gray')
plt.title('Median Filter (3x3)')
plt.axis('off')

midpoint_filtered = generic_filter(gray, midpoint_filter, size=3)
plt.subplot(3, 3, 9)
plt.imshow(midpoint_filtered, cmap='gray', vmin=0, vmax=255)
plt.title('Midpoint Filter (3x3)')
plt.axis('off')

plt.tight_layout()
plt.show()
```
