```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image_path = 'ahmedyassertaha.jpg' 
image_rgb = cv2.imread(image_path)
```

```python
image_rgb = cv2.cvtColor(image_rgb, cv2.COLOR_BGR2RGB)
```

```python
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(image_rgb)
plt.title('RGB Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.text(0.1, 0.5, f'Shape: {image_rgb.shape}', fontsize=12, color='black')
plt.axis('off')

plt.show()
```

```python
gray_image = cv2.cvtColor(image_rgb, cv2.COLOR_BGR2GRAY)
```

```python
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(gray_image, cmap='gray')
plt.title('Grayscale Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.text(0.1, 0.5, f'Shape: {gray_image.shape}', fontsize=12, color='black')
plt.axis('off')

plt.show()
```

```python
_, binary_image = cv2.threshold(gray_image, 127, 255, cv2.THRESH_BINARY)

plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(binary_image, cmap='gray')
plt.title('Binary-Scale Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.text(0.1, 0.5, f'Shape: {binary_image.shape}', fontsize=12, color='black')
plt.axis('off')

plt.show()
```
